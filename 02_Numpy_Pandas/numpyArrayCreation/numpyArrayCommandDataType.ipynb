```python
#Create a simple List
import numpy as np
```

```python
n=np.array([1,2,3,4,5])
print(n,n.dtype)
n1=np.array([1,2,3,4,13.5])
print(n1,n1.dtype)
```

```python
#We can aslo have an Object Type in numpy
#which is not prefered because numpy is Optimize for numeric computations
#we can store thought Object data in numby Array but its not Preferable
```

 For String and Object Type Data we dont prefer the Numpy

```python
#Here is the example of The Object Type Data In the Numpy Array
#When we store the homogenious Data Python Automaticly convert it into Object Type

n3=np.array(["John",{1,2,3},13.13])
print(n3,n3.dtype)
```

```python
#We aslo have a complexType
x=np.array([3+4j])
y=np.array([5+3j])
w=x+y
print(w,w.dtype)
print(type(w))
```
