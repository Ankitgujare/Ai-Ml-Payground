```python
df
```

```python
# get Only Duplicate Rows 
df[df.duplicated()]
```

```python
df[df["name"].duplicated()]
```

```python
# Get All Duplicates (including first occurrence)
df[df.duplicated(keep=False)]
```

```python
    # Check Row-wise vs Column-wise Summary
    #Row-wise duplicates count:
    
    df.duplicated().sum()
```

```python
 # Column wise Duplicate Count 
df.T.duplicated().sum()
```

```python
Case	
Row-wise duplicates
df.duplicated()
Specific column	
df.duplicated(subset=['col'])
Column values	
df['col'].duplicated()
Column-wise duplicates
df.T.duplicated()
Get duplicate rows	
df[df.duplicated()]
Get all duplicates
df[df.duplicated(keep=False)]
```
