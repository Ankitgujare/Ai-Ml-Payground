#### there we Diffrent Ways to create a Dataframe
- from Dictionary
- from List
- From Numpy Array 

lets see them One By One

```python
import pandas as pd 
import numpy as np 
```

```python
#From Dictionary

info={
    "name":["Aman","Rutuja","Rahul","Pooja"],
    "Age":[22,44,12,33],
    "RollNumber":[1,2,3,4]
}


df=pd.DataFrame(info)
print(df)
df
df.index
df.columns
```

```python
#2.with the list of List
student=[["aman","rahul","pooja"],[1,2,3],["male","Male","Female"]]
df1=pd.DataFrame(student)
df1
print(df1)
```

```python
#with numpy Array
user=np.array([["Aman","Pooja","Swapnil"],[1,2,3],["M","F","M"]])
df3=pd.DataFrame(user,index=["Name","RollNumber","Gender"])
df3
```
