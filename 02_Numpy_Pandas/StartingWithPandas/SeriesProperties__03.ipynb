#### Pandas series(Properties)
- homogenious
- Vectorize Operations
- Handling missing Values with NAN
- Mutable values but immutable size

```python
#why we say series in homogenious
s1=pd.Series([12,24,44,55,32],index=["Aman","rohan","Sushil","Pooja","Kaka"])
print(s1)
```

```python
#But suppose if we have values Like
s2=pd.Series([12,24,44,55,"Age"],index=["Aman","rohan","Sushil","Pooja",24])
print(s2)
```

You can see Pandas Automaticaly Converted it into type Object as we have a value of diffrent Type

we can aslo do Vectorized Operations with series which in the end make 
Operations efficient (faster) and in the memory Optimization too (Dosent Occupy an extra memory)

s3=pd.Series([1,2,3,4,5,6])
s4=pd.Series([12,22,32,42,52,62])
print(s3+s4)

Mutable values but immutable size
in series we can we can change the values but addition or removal of element dosent affect the Original size of Series Instead it creats new Series everytime we do that

```python
s5=pd.Series([12,22,32,42,52,62])
s5[2]=999
print("Original series",s5)
print("after changing",s5)

```

```python
#we add or removed the value
s5=pd.Series([12,22,32,42,52,62])
s6=s5.drop(1)
s6
```

```python
s7=s6.drop(4)
s7
```
