#### Vectorization and BroadCasting

```python
 import numpy as np
```

```python
#Broadcasting
arr=np.array([1,2,3,4,5])
arr1=np.array([5])
print(arr+arr1)
```

#### we use BroadCasting when we are using Numpy array for an Operation and having Diffrences In Size
### You can See we are doing Array Addition

And  Other One is (5,)

In this case we are doing an Addition
how its Possible to add such array which is not compatible to add 
With the help of the BroadCasting 

it just scale the [5] into [5,5,5,5,5] & And then it add two Array

When operating on two arrays, NumPy compares their shapes element-wise. It starts with the trailing (i.e. rightmost) dimension and works its way left. Two dimensions are compatible when

they are equal, or

one of them is 1.

```python
a=np.array([1,2,3,4,5])
a1=np.array([[1,2,3,4,5],[2,2,2,2,2]])
print(a,a.shape)
print(a1,a1.shape)
print(a+a1)
```

```python
b=np.array([[1,2,3,4,5],[2,2,2,2,2],[3,3,3,3,3]])
b1=np.array([[1,2,3,4,5],[2,2,2,2,2]])
print(b,b.shape)
print(b1,b1.shape)
print(b+b1)
```
