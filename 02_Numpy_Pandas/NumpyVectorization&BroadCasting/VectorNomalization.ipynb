```python
import numpy as np
```

```python
arr=np.array([[1,2],[3,4]])
print(arr,arr.shape)
```

```python
#Calculate the mean value first
means=np.mean(arr)
std_value=np.std(arr)
print(std_value)
print(means)
```

```python
#Normalize an array
normalizedArray=(arr-means)/std_value
```

```python
print(normalizedArray)
```

# Now the normalized Array we have 
- the mean of the array =0
- standered Deviation values is as close as 1 as Possible
