```python
import numpy as np
```

```python
#Aggregation Functions
arr=np.array([1,2,3,4,5])
print(np.sum(arr))
print(np.prod(arr))
print(np.max(arr))
print(np.min(arr)) 
print(np.sqrt(arr))
print(np.var(arr))          #Returns variance
print(np.mean(arr))         #Returns average value
print(np.median(arr))
print(np.argmin(arr))       #returns the index of min value       
print(np.argmax(arr))       #returns the index of max value 
print(np.square(arr))       #returns the Square of an array
print(pow(10,2))            #not a numpy function return a to power b
```

```python
#Logs
print(np.log(arr))    #Natural logarithm (base e)
print(np.log10(arr))    #Natural logarithm (base10)
print(np.log2(arr))    #Natural logarithm (base 2)
print(np.exp(arr))    #Exponential (e^x)
```

```python
#Rounding
print(np.round(3.14))   #Rounds to nearest value
print(np.floor(2.88))   #Rounds to down
print(np.floor(2.88))   #Rounds to up`
print(np.trunc(3.14))   #Truncates decimal part
print(np.unique(arr))   #return Unique Element
print(np.sort(arr))    #return Unique Element
print(np.abs(arr))     #return absolute Value


```
